```python
import json
import requests as r
```

```python
with open('/Users/jtamras/API_Keys/college_api_key.txt') as f:
    api_key = f.read().replace('\n', '')
    #print(api_key)

with open('/Users/jtamras/API_Keys/google_maps_key.txt') as f:
    maps_key = f.read().replace('\n', '')
    #print(maps_key)
```

```python
def params(api_key):
    """
    Construct url, headers and url_params. Reference API docs (link above) to use the arguments
    """
    # [YOUR CODE HERE]
    # What is the url endpoint for search?
    url = f'https://api.data.gov/ed/collegescorecard/v1/schools.json?fields=school.name,latest.student.size&sort=latest.student.size:desc&per_page=100'
    # SPACES in url is problematic. How should you handle location containing spaces?
    url_params = {'api_key': api_key}

    return url, url_params


p = params(api_key=api_key)

universities_raw = r.get(p[0],p[1]).content
top_univ = json.loads(universities_raw.decode('utf-8'))['results']

top_univ
names = []
for c in top_univ:
    names.append(c['school.name'])

names
```

```python
def maps_params(place_id):
    """
    Construct url, headers and url_params. Reference API docs (link above) to use the arguments
    """
    # [YOUR CODE HERE]
    # What is the url endpoint for search?
    url = 'https://maps.googleapis.com/maps/api/place/details/json'
    # SPACES in url is problematic. How should you handle location containing spaces?
    url_params = {'key': maps_key, 'fields' : 'formatted_address', 'place_id' : place_id}

    return url, url_params

def maps_places_params(name):
    url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json'
    url_params = { 'key': maps_key, 'inputtype' : 'textquery', 'input': name}
    return url, url_params

def getUnivAddress(place_id):
    url, params = maps_params(place_id)
    result = r.get(url, params=params).content
    result = json.loads(result.decode('utf-8'))
    return result['result']['formatted_address']

def getPlaceId(univ_name):
    url, url_params = maps_places_params(univ_name)
    result = r.get(url, params=url_params).content
    result = json.loads(result.decode('utf-8'))
    return result['candidates'][0]['place_id']

university_names = []

for u in top_univ:
    university_names.append(u['school.name'])

university_addresses = []

for n in university_names:
    place_id = getPlaceId(n)
    university_addresses.append({'univ_name' : n, 'address' :getUnivAddress(place_id)})
    
print(university_addresses)
print(len(university_addresses))
```
